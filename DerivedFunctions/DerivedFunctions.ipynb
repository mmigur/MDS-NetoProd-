```python
import sympy
import numpy as np
import matplotlib.pyplot as plt
```

## Задание 1
Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$

В точке x = 1

```python
x = sympy.Symbol('x')
function = x ** 5 + 4 * sympy.sin(2 * x) + sympy.cos(3 * x + 3)
```

```python
first_derivative_function = sympy.diff(function, x)
second_derivative_function = sympy.diff(first_derivative_function, x)
```

```python
print('Значение первой производной функции ' + str(sympy.diff(first_derivative_function).evalf(subs={'x': 1})))
print('Значение второй производной функции ' + str(sympy.diff(second_derivative_function).evalf(subs={'x': 1})))
```

## Задание 2
Постройте график первой и второй производной функции

$f(x) = sin^5(2x+1)$

На отрезке [-5; 5].

```python
x = sympy.Symbol('x')
function = sympy.sin(2 * x + 1) ** 5
```

```python
input_range = np.arange(-5, 5, 0.1)
```

```python
first_derivative = sympy.diff(function, x)
output_values_fisrt = [first_derivative.subs(x, i) for i in input_range]
```

```python
second_derivative = sympy.diff(first_derivative, x)
output_values_second = [second_derivative.subs(x, i) for i in input_range]
```

```python
plt.title('Производные двух функций на промежутке [-5, 5]')
plt.plot(input_range, output_values_fisrt)
plt.plot(input_range, output_values_second)
plt.xlim([-5, 5])
plt.ylim([-5, 5])
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['Первая производная', 'Вторая производная'])
```
