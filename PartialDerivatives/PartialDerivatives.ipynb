```python
import numpy as np
import sympy
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
```

## Задание 1
Посчитайте частные производные функций:

1) $f(x,y)=2x^2y^3 + 1/x + y^2x + 7$

2) $f(x,y)=x^2y - sin(xy) + cos(x^2) + 6y$

```python
x = sympy.Symbol('x')
y = sympy.Symbol('y')
```

```python
function_one = (2 * x ** 2 * y ** 3) + (1 / x) + (y ** 2 * x) + 7
function_two = (x ** 2 * y) - sympy.sin(x*y) + sympy.cos(x ** 2) + 6 * y
```

```python
print(f'Частная производная по x для первой функции - {sympy.diff(function_one, x, x)}')
print(f'Частная производная по y для первой функции - {sympy.diff(function_one, x, y)}')
```

Частная производная по x для первой функции = $2(2y^3 + x^(-3))$

Частная производная по y для первой функции = $2y(6xy + 1)$

```python
print(f'Частная производная по x для второй функции - {sympy.diff(function_two, x, x)}')
print(f'Частная производная по y для второй функции - {sympy.diff(function_two, x, y)}')
```

Частная производная по x для второй функции = $-4x^2cos(x^2) + y^2sin(xy) + 2y - 2sin(x^2)$

Частная производная по y для второй функции = $xy * sin(xy) + 2x - cox(xy)$

## Задание 2
Вычислите градиент функции:

$f(x,y)=2x^2y^3 + 1/x + y^2x + 7$

в точке $(1,2)$

```python
x = sympy.Symbol('x')
y = sympy.Symbol('y')
```

```python
function = 2 * x ** 2 * y ** 3 + 1 / x + y ** 2 * x + 7
```

```python
x_derivative = sympy.diff(function, x, x)
y_derivative = sympy.diff(function, x, y)
```

```python
grad_f = (x_derivative.subs(x, 1).subs(y, 2), y_derivative.subs(x, 1).subs(y, 2))
```

```python
print(f'Градиент нашей функции в точке (1, 2) = {grad_f}')
```

## Задание 3
Постройте трехмерный график функции

$f(x)=x_2^2-x_1^2 x_2 + 2 x_1 x_2$

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

# Данные
X1 = np.arange(-5, 5, 0.25)
X2 = np.arange(-5, 5, 0.25)

X1, X2 = np.meshgrid(X1, X2)
Z = (X2 ** 2) - (X1 ** 2 * X2) + 2 * X1 * X2


# График поверхности
surf = ax.plot_surface(X1, X2, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Настройка оси Z
ax.set_zlim(Z.min(), Z.max())
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# Цветовая легенда
fig.colorbar(surf, shrink=0.5, aspect=5)

ax.azim = 15
ax.dist = 10
ax.elev = 30

plt.show()
```
