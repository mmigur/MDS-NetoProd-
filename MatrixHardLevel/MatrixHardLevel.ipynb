```python
import numpy as np
import matplotlib.pyplot as plt
```

## Задание 1
Дан объект в 2D пространстве

При помощи линейных отображений:

* Уменьшить объект в два раза
* Повернуть на 130 градусов
* Отразить объект относительно прямой y=x (поможет матрица [[0, 1, 0], [1, 0, 0], [0, 0, 1]]

```python
def show_matrix(matrix):
    x = matrix[:,0]
    y = matrix[:,1]
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.ylim([-200, 200])
    plt.xlim([-200, 200])
```

```python
A = np.array([
    [-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]
])

show_matrix(A)
```

### Уменьшение 

```python
MATRIX = np.array([
    [-100, -100, 1],
    [-100, 100, 1],
    [100, 100, 1],
    [0, 0, 1],
    [100, -100, 1],
    [-100, -100, 1]
])

SCALE = 0.5
SCALE_MATRIX = np.array([
    [SCALE, 0, 0], 
    [0, SCALE, 0],
    [0, 0, SCALE]
])

result_scaled_matrix = np.array([np.matmul(SCALE_MATRIX, vector) for vector in MATRIX])
show_matrix(result_scaled_matrix)
```

### Поворот

```python
ANGLE_OF_ROTATION = (130 * np.pi) / 180

rotation_matrix = np.array([
    [np.cos(ANGLE_OF_ROTATION), -np.sin(ANGLE_OF_ROTATION),0],
    [np.sin(ANGLE_OF_ROTATION), np.cos(ANGLE_OF_ROTATION), 0],
    [0, 0, 1]
])

rotated_matrix = np.array([np.matmul(vector, rotation_matrix) for vector in result_scaled_matrix])
```

```python
show_matrix(rotated_matrix)
```

### Отражение

```python
REFLECTION_MATRIX = np.array([
    [0, -1, 0], 
    [-1, 0, 0], 
    [0, 0, 1]
])

reflected_matrix = np.array([np.matmul(REFLECTION_MATRIX, vector) for vector in rotated_matrix])
show_matrix(reflected_matrix)
```

## Задание 2
Найдите спектральное разложение матрицы:

```python
# MATRIX = S * D * S.T

MATRIX = np.array([
    [1, 2],
    [2, 3]
])
```

```python
EIGEN_VECTORS = np.linalg.eig(MATRIX)[1]
```

```python
DIAGONAL_MATRIX = np.zeros((2, 2))

count = 0
for i in range(len(DIAGONAL_MATRIX)):
    DIAGONAL_MATRIX[i, count] = MATRIX.diagonal()[i]
    count += 1
```

```python
EIGEN_VECTORS_TRANSPOSE = np.transpose(EIGEN_VECTORS)
```

### Спектральное разложение

```python
np.matmul(EIGEN_VECTORS, DIAGONAL_MATRIX, EIGEN_VECTORS_TRANSPOSE)
```
