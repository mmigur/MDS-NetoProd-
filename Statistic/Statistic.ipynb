```python
import pandas as d
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import expon
%matplotlib inline
```

## Задание.
В этом задании нужно убедиться, что ЦПТ действительно работает.

Создайте случайную величину из любого выбранного вами распределения, для разных значений n сгенерируйте 1000 выборок размера n. Постройте гистрограммы средних этих выборок и сделайте выводы.

```python
rv = expon(loc=0, scale=2)
sample = rv.rvs(size=1000)
```

```python
x = np.linspace(0,2,100)
pdf = rv.pdf(x)
plt.hist(sample, density=True, bins=30, label='sample histogram')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5)
plt.legend()
plt.ylabel('fraction of samples, f(x)')
plt.xlabel('x')
plt.show()
```

```python
SAMPLES_SIZE = 30

samples_mean = np.array([])

SAMPLES_COUNT = 1000

for number in range(SAMPLES_COUNT): 
    current_sample = rv.rvs(size=SAMPLES_SIZE)
    samples_mean = np.append(samples_mean, current_sample.mean())
    
plt.hist(samples_mean, density=True, label='sample of ' + str(SAMPLES_SIZE) + ' histogram')
plt.xlabel('x')
plt.ylabel('samples means')
plt.legend()
plt.show()
```
