## Задание 1
В магазин привезли устройства с 3-х разных предприятий.

Соотношение устройств следующее: 20% - продукция первого предприятия, 30% - продукция второго предприятия, 50% - продукция третьего предприятия; далее, 10% продукции первого предприятия высшего сорта, на втором предприятии - 5% и на третьем - 20% продукции высшего сорта.

Найти вероятность того, что случайно купленная новая продукция окажется высшего сорта.

###  Решение
P(A) - Новая купленная техника окажется первого сорта.

* 20% - 1 (высший сорт - 10%)
* 30% - 2 (высший сорт - 5%)
* 50% - 3 (высший сорт - 20%)

Пусть событие A - покупка новой продукции высшего сорта. Тогда, применяя формулу полной вероятности, получим:

P(A) = P(A|1) * P(1) + P(A|2) * P(2) + P(A|3) * P(3),

где P(1), P(2), P(3) - вероятности покупки продукции первого, второго и третьего предприятий соответственно.

P(1) = 0.2, P(2) = 0.3, P(3) = 0.5.

P(A|1), P(A|2), P(A|3) - вероятности того, что продукция высшего сорта при покупке продукции первого, второго и третьего предприятий соответственно.

P(A|1) = 0.1, P(A|2) = 0.05, P(A|3) = 0.2.

Подставляя данные в формулу полной вероятности, получаем:

P(A) = 0.1 * 0.2 + 0.05 * 0.3 + 0.2 * 0.5 = 0.02 + 0.015 + 0.1 = 0.135.

Ответ: вероятность покупки новой продукции высшего сорта равна 13.5%.

## Задание 3
Изучите 3 распределения из библиотеки scipy: Нормальное, Экспоненциальное, Стьюдента. Используя функцию плотности распредления (pdf), постройте пять графиков плотностей для каждого распределения при разных параметрах. Запишите ваши наблюдения: как и на что влияют параметры у каждого распределения.

## Нормальное распределение

```python
x_norm = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
```

```python
plt.plot(x_norm, norm.pdf(x_norm), 'y', alpha=0.5, label='norm pdf(scale=1)')
plt.plot(x_norm, norm.pdf(x_norm, scale=2), 'g', alpha=0.5, label='norm pdf(scale=2)')
plt.plot(x_norm, norm.pdf(x_norm, scale=3), 'r', alpha=0.5, label='norm pdf(scale=3)')
plt.plot(x_norm, norm.pdf(x_norm, scale=4), 'b', alpha=0.5, label='norm pdf(scale=4)')
plt.plot(x_norm, norm.pdf(x_norm, scale=5), 'p', alpha=0.5, label='norm pdf(scale=5')
plt.legend()
plt.show()
```

## Экспоненциальное распределение

```python
x_exp = np.linspace(expon.ppf(0.01), expon.ppf(0.99), 100)
```

```python
plt.plot(x_exp, expon.pdf(x_exp, scale=5), 'y', alpha=0.5, label='expon pdf(lambda=5)')
plt.plot(x_exp, expon.pdf(x_exp, scale=6), 'g', alpha=0.5, label='expon pdf(lambda=6)')
plt.plot(x_exp, expon.pdf(x_exp, scale=7), 'r', alpha=0.5, label='expon pdf(lambda=7)')
plt.plot(x_exp, expon.pdf(x_exp, scale=8), 'b', alpha=0.5, label='expon pdf(lambda=8)')
plt.plot(x_exp, expon.pdf(x_exp, scale=9), 'p', alpha=0.5, label='expon pdf(lambda=9)')
plt.legend()
plt.show()
```

## Распределение Стьюдента

```python
x_t = np.linspace(t.ppf(0.01, 2.71), t.ppf(0.99, 2.71), 100)
```

```python
plt.plot(x_t, t.pdf(x_t, scale=10, df=2.71), 'y', alpha=0.5, label='t pdf(n=10)')
plt.plot(x_t, t.pdf(x_t, scale=11, df=2.71), 'g', alpha=0.5, label='t pdf(n=11)')
plt.plot(x_t, t.pdf(x_t, scale=12, df=2.71), 'r', alpha=0.5, label='t pdf(n=12)')
plt.plot(x_t, t.pdf(x_t, scale=13, df=2.71), 'b', alpha=0.5, label='t pdf(n=13)')
plt.plot(x_t, t.pdf(x_t, scale=14, df=2.71), 'p', alpha=0.5, label='t pdf(n=14)')
plt.legend()
plt.show()
```
